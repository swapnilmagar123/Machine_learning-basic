```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("Social_Network_Ads.csv")
df.head()
from sklearn.model_selection import train_test_split
```

```python
df=df.iloc[:,2:]
df

```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,2:3]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
scalar = sc.fit(x_train)
```

```python
X_train_scaled = sc.transform(x_train)
X_test_scaled = sc.transform(x_test)
```

```python
# 1. fit() – Learning the Pattern
# Purpose:
# fit() is used to train or fit a model to the data. It learns the parameters (like mean, standard deviation, or weights) from the data but does not transform or modify the data.

 # Output:
# Returns the model/transformer itself, ready to apply further transformations or predictions.


# 2. fit_transform() – Learning and Transforming
# Purpose:
# fit_transform() combines the steps of fitting and transforming the data in one go. It fits the model to the data (like fit()) and immediately transforms the data using the learned parameters.

# Output:
# Returns the transformed data directly.
```
